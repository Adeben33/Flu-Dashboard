```python
# df_weekly = df_single['Cases'].asfreq('W')
df_weekly = df_single['Cases'].asfreq('W-SAT')
```

```python
# --- Ensure the data is numeric and handle any non-numeric values ---
df_weekly = pd.to_numeric(df_weekly, errors='coerce')  # Convert to numeric, invalid entries become NaN

# --- Fill missing values ---
df_weekly = df_weekly.interpolate(method='linear')  # You can choose another method if necessary

# --- Log-transform the data (with small offset to avoid log(0)) ---
df_log = np.log(df_weekly + 1e-6)  # Add a small offset to avoid log(0)

# --- STL decomposition on the log-transformed data ---
stl_log = STL(df_log, period=52)
result_log = stl_log.fit()

# --- Inverse transform components (exp) ---
observed = np.exp(result_log.observed)
trend = np.exp(result_log.trend)
seasonal = np.exp(result_log.seasonal)

# --- Plot the results ---
plt.figure(figsize=(12, 6))
plt.plot(df_weekly.index, df_weekly, label='Original Cases', marker='o', alpha=0.5)
plt.plot(df_weekly.index, seasonal * trend, label='Exp(Seasonal + Trend)', color='red')
plt.title('STL Decomposition (on log scale, re-transformed)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt

# Filter and prepare time series data for one virus (Influenza)
df_single = df[df['Virus'] == 'Influenza'].sort_index()

# Ensure a complete weekly time index (resample)
# df_weekly = df_single['Cases'].asfreq('W')
df_weekly = df_single['Cases'].asfreq('W-SAT')

print(df_weekly)

# Perform additive and multiplicative decompositions
additive_result = seasonal_decompose(df_weekly, model='additive', period=42)
multiplicative_result = seasonal_decompose(df_weekly, model='multiplicative', period=42)

# Plotting
fig, axes = plt.subplots(2, 1, figsize=(14, 10), sharex=True)

# Additive Plot
axes[0].plot(df_weekly.index, df_weekly, label='Observed', marker='o', alpha=0.6)
axes[0].plot(df_weekly.index, additive_result.seasonal, label='Additive Seasonality', linestyle='--', color='orange')
axes[0].set_title('Additive Decomposition: Observed vs Seasonal')
axes[0].legend()
axes[0].grid(True)

# Multiplicative Plot
axes[1].plot(df_weekly.index, df_weekly, label='Observed', marker='o', alpha=0.6)
# Multiply seasonal and trend (avoids plotting component alone)
product = multiplicative_result.seasonal * multiplicative_result.trend
axes[1].plot(df_weekly.index, product, label='Multiplicative Seasonality × Trend', linestyle='--', color='green')
axes[1].set_title('Multiplicative Decomposition: Observed vs Seasonal × Trend')
axes[1].legend()
axes[1].grid(True)

plt.tight_layout()
plt.show()
```

# Fourier Analysis

```python
df_single = df[df['Virus'] == 'Influenza'].sort_index()
df_weekly = df_single['Cases'].asfreq('W-SAT')
df_weekly = df_weekly.interpolate(method='linear')
```

```python
def add_fourier_terms(df, period=52, order=2):
    t = np.arange(len(df))
    for k in range(1, order + 1):
        df[f'sin_{k}'] = np.sin(2 * np.pi * k * t / period)
        df[f'cos_{k}'] = np.cos(2 * np.pi * k * t / period)
    return df

df_fourier = df_weekly.to_frame(name='Cases')
df_fourier = add_fourier_terms(df_fourier, period=52, order=2)
```

```python
import statsmodels.api as sm

# Build design matrix with Fourier terms
X = df_fourier[[col for col in df_fourier.columns if 'sin' in col or 'cos' in col]]
X = sm.add_constant(X)  # Add intercept
y = df_fourier['Cases']

# Fit model
model = sm.OLS(y, X).fit()
df_fourier['Fitted'] = model.predict(X)
```

```python
plt.figure(figsize=(12, 5))
plt.plot(df_fourier.index, df_fourier['Cases'], label='Observed Cases', marker='o', alpha=0.6)
plt.plot(df_fourier.index, df_fourier['Fitted'], label='Fourier Fitted Seasonality', color='red')
plt.title('Fourier Series Seasonal Fit')
plt.xlabel('Date')
plt.ylabel('Cases')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# --- Step 1: Prepare Weekly Influenza Cases ---
df_single = df[df['Virus'] == 'Influenza'].sort_index()
df_weekly = df_single['Cases'].asfreq('W-SAT')
df_weekly = df_weekly.interpolate(method='linear')

# --- Step 2: Add Fourier Terms + Trend ---
def add_fourier_terms(df, period=52, order=4):
    t = np.arange(len(df))
    for k in range(1, order + 1):
        df[f'sin_{k}'] = np.sin(2 * np.pi * k * t / period)
        df[f'cos_{k}'] = np.cos(2 * np.pi * k * t / period)
    df['time'] = t  # Add trend component
    return df

df_fourier = df_weekly.to_frame(name='Cases')
df_fourier = add_fourier_terms(df_fourier, period=52, order=2)

# --- Step 3: Fit OLS Model ---
X_cols = [col for col in df_fourier.columns if 'sin' in col or 'cos' in col] + ['time']
X = sm.add_constant(df_fourier[X_cols])
y = df_fourier['Cases']

model = sm.OLS(y, X).fit()
df_fourier['Fitted'] = model.predict(X)
df_fourier['Residual'] = df_fourier['Cases'] - df_fourier['Fitted']

# --- Step 4: Plot Fitted vs Observed ---
plt.figure(figsize=(12, 5))
plt.plot(df_fourier.index, df_fourier['Cases'], label='Observed Cases', marker='o', alpha=0.6)
plt.plot(df_fourier.index, df_fourier['Fitted'], label='Fourier Fitted Seasonality + Trend', color='red')
plt.title('Fourier Series Fit with Trend')
plt.xlabel('Date')
plt.ylabel('Flu Cases')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# --- Step 5: Plot Residuals ---
plt.figure(figsize=(12, 4))
plt.plot(df_fourier.index, df_fourier['Residual'], label='Residuals', color='gray')
plt.axhline(0, linestyle='--', color='black')
plt.title('Residuals After Fourier + Trend Fit')
plt.xlabel('Date')
plt.grid(True)
plt.tight_layout()
plt.legend()
plt.show()

# --- Optional: Print summary ---
print(model.summary())
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

# --- Step 1: Prepare Weekly Influenza Data ---
df_single = df[df['Virus'] == 'Influenza'].sort_index()
df_weekly = df_single['Cases'].asfreq('W-SAT')
df_weekly = df_weekly.interpolate(method='linear')

# --- Step 2: Add Fourier Terms + Trend Function ---
def add_fourier_terms(df, period=52, order=2, start_t=0):
    t = np.arange(start_t, start_t + len(df))
    for k in range(1, order + 1):
        df[f'sin_{k}'] = np.sin(2 * np.pi * k * t / period)
        df[f'cos_{k}'] = np.cos(2 * np.pi * k * t / period)
    df['time'] = t
    return df

# --- Step 3: Fit Model to Historical Data ---
df_fourier = df_weekly.to_frame(name='Cases')
df_fourier = add_fourier_terms(df_fourier, period=52, order=2)

X_cols = [col for col in df_fourier.columns if 'sin' in col or 'cos' in col] + ['time']
X = sm.add_constant(df_fourier[X_cols])
y = df_fourier['Cases']

model = sm.OLS(y, X).fit()
df_fourier['Fitted'] = model.predict(X)

# --- Step 4: Forecast Future Weeks (e.g., next 26 weeks) ---
n_forecast = 26
future_dates = pd.date_range(df_weekly.index[-1] + pd.Timedelta(weeks=1), periods=n_forecast, freq='W-SAT')
df_future = pd.DataFrame(index=future_dates)
df_future = add_fourier_terms(df_future, period=52, order=2, start_t=len(df_fourier))

X_future = sm.add_constant(df_future[X_cols])
df_future['Forecast'] = model.predict(X_future)

# --- Step 5: Plot Historical + Forecasted Flu Cases ---
plt.figure(figsize=(14, 6))
plt.plot(df_fourier.index, df_fourier['Cases'], label='Observed', marker='o', alpha=0.6)
plt.plot(df_fourier.index, df_fourier['Fitted'], label='Fitted', color='red')
plt.plot(df_future.index, df_future['Forecast'], label='Forecast (Next 26 Weeks)', color='green', linestyle='--', marker='o')
plt.axvline(df_fourier.index[-1], color='gray', linestyle='--', label='Forecast Start')
plt.title('Flu Cases Forecast Using Fourier Series + Trend')
plt.xlabel('Date')
plt.ylabel('Flu Cases')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.statespace.sarimax import SARIMAX
from prophet import Prophet
from statsmodels.tsa.seasonal import STL
from statsmodels.tsa.forecasting.stl import STLForecast

# --- Step 1: Prepare Weekly Influenza Data ---
df_single = df[df['Virus'] == 'Influenza'].sort_index()
df_weekly = df_single['Cases'].asfreq('W-SAT')
df_weekly = df_weekly.interpolate(method='linear')

# --- Step 2: Add Fourier Terms + Trend Function ---
def add_fourier_terms(df, period=52, order=2, start_t=0):
    t = np.arange(start_t, start_t + len(df))
    for k in range(1, order + 1):
        df[f'sin_{k}'] = np.sin(2 * np.pi * k * t / period)
        df[f'cos_{k}'] = np.cos(2 * np.pi * k * t / period)
    df['time'] = t
    return df

# --- Step 3: Fit Fourier Model to Historical Data ---
df_fourier = df_weekly.to_frame(name='Cases')
df_fourier = add_fourier_terms(df_fourier, period=52, order=2)

X_cols = [col for col in df_fourier.columns if 'sin' in col or 'cos' in col] + ['time']
X = sm.add_constant(df_fourier[X_cols])
y = df_fourier['Cases']

model = sm.OLS(y, X).fit()
df_fourier['Fitted'] = model.predict(X)

# --- Step 4: Forecast Future Weeks (e.g., next 26 weeks) ---
n_forecast = 26
future_dates = pd.date_range(df_weekly.index[-1] + pd.Timedelta(weeks=1), periods=n_forecast, freq='W-SAT')
df_future = pd.DataFrame(index=future_dates)
df_future = add_fourier_terms(df_future, period=52, order=2, start_t=len(df_fourier))

X_future = sm.add_constant(df_future[X_cols])
df_future['Forecast'] = model.predict(X_future)

# --- Step 5: SARIMA Forecasting ---
sarima_model = SARIMAX(df_weekly, order=(1,1,1), seasonal_order=(1,1,1,52))
sarima_result = sarima_model.fit()

sarima_forecast = sarima_result.get_forecast(steps=n_forecast)
forecast_mean = sarima_forecast.predicted_mean
conf_int = sarima_forecast.conf_int()

# --- Step 6: Prophet Forecasting ---
df_prophet = df_weekly.reset_index()
df_prophet.columns = ['ds', 'y']

prophet_model = Prophet(weekly_seasonality=True)
prophet_model.fit(df_prophet)

future = prophet_model.make_future_dataframe(periods=n_forecast, freq='W-SAT')
forecast = prophet_model.predict(future)

# --- Step 7: STL Forecasting ---
stl_forecast = STLForecast(df_weekly, ARIMA, model_kwargs={"order": (1,1,1)}, period=52)
stl_result = stl_forecast.fit()
stl_forecast_values = stl_result.forecast(n_forecast)

# --- Step 8: Combine Forecasts into one DataFrame ---
forecast_df = pd.DataFrame(index=df_future.index)
forecast_df['Fourier'] = df_future['Forecast']
forecast_df['STL'] = stl_forecast_values
forecast_df['SARIMA'] = forecast_mean
forecast_df['Prophet'] = forecast.set_index('ds').loc[forecast_df.index, 'yhat']

# --- Step 9: Plot All Forecasts ---
plt.figure(figsize=(16, 6))
plt.plot(df_weekly, label='Observed', marker='o', alpha=0.6)

for col in forecast_df.columns:
    plt.plot(forecast_df.index, forecast_df[col], label=f'{col} Forecast', linestyle='--')

plt.axvline(df_weekly.index[-1], color='gray', linestyle='--', label='Forecast Start')
plt.title('Observed Flu Cases vs Forecasts from Four Methods')
plt.xlabel('Date')
plt.ylabel('Flu Cases')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# --- IMPORT LIBRARIES ---
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.forecasting.stl import STLForecast
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.seasonal import STL
from cmdstanpy import CmdStanModel
import streamlit as st
import pydeck as pdk
from pandas.plotting import register_matplotlib_converters
import os

register_matplotlib_converters()

st.set_page_config(layout="wide")
st.title("🦠 Influenza in Canada")

# --- SIDEBAR ---
st.sidebar.header("Filter")
date_range = st.sidebar.date_input("Select Date Range")
region = st.sidebar.selectbox("Region", ["Canada (National)", "Ontario", "Quebec", "BC", "Alberta"])
n_weeks = st.sidebar.slider("Weeks to Forecast", min_value=4, max_value=52, value=26)
season = st.sidebar.selectbox("Season", ["2023-2024", "2024-2025"])

uploaded_file = st.sidebar.file_uploader("Upload CSV (with surveillance data)", type='csv')

# --- MAIN WORK ---
if uploaded_file:
    df = pd.read_csv(uploaded_file)

    # Create Date
    df['Date'] = pd.to_datetime(df['Year'].astype(str) + df['Surveilaince Week'].astype(str) + '1', format='%G%V%u')
    df.set_index('Date', inplace=True)

    # Influenza A and Total
    df['Influenza A'] = df['A(H1N1)'] + df['A(H3N2)'] + df['A(Sub typed)']
    df['Influenza Total'] = df['Influenza A'] + df['Influenza B']

    # Extract key columns
    ah1n1 = df['A(H1N1)']
    ah3n2 = df['A(H3N2)']
    bunsubtyped = df['A(Sub typed)']
    flu_b = df['Influenza B']
    influenza_total = df['Influenza Total']
    perc_a = df['Percent Positive A']
    perc_b = df['Percent Positive B']

    # --- KPIs ---
    total_cases = int(ah1n1.sum() + ah3n2.sum() + bunsubtyped.sum() + flu_b.sum())
    last_week = df.index.max()
    current_week_data = df.loc[df.index == last_week]
    weekly_cases = int(current_week_data[['A(H1N1)', 'A(H3N2)', 'A(Sub typed)', 'Influenza B']].sum(axis=1).values[0])

    avg_perc_a = round(perc_a.mean(), 2)
    avg_perc_b = round(perc_b.mean(), 2)
    latest_perc_a = round(current_week_data['Percent Positive A'].values[0], 2)
    latest_perc_b = round(current_week_data['Percent Positive B'].values[0], 2)

    kpi1, kpi2, kpi3, kpi4, kpi5, kpi6 = st.columns(6)
    kpi1.metric("Total Cases", f"{total_cases:,}")
    kpi2.metric("New Cases (Recent Week)", f"{weekly_cases:,}")
    kpi3.metric("Avg % Positive A", f"{avg_perc_a}%")
    kpi4.metric("Avg % Positive B", f"{avg_perc_b}%")
    kpi5.metric("This Week % Positive A", f"{latest_perc_a}%")
    kpi6.metric("This Week % Positive B", f"{latest_perc_b}%")

    # --- Plots ---
    col1, col2 = st.columns(2)

    # --- Influenza Composition ---
    with col1:
        st.markdown("### Influenza Activity Overview")
        fig1, ax1 = plt.subplots(figsize=(11.5, 6), dpi=150)
        ax1.bar(df.index, ah1n1, label='A(H1N1)', color='firebrick')
        ax1.bar(df.index, ah3n2, bottom=ah1n1, label='A(H3N2)', color='lightcoral')
        ax1.bar(df.index, bunsubtyped, bottom=ah1n1 + ah3n2, label='A(Unsubtyped)', color='mistyrose', hatch='//')
        ax1.bar(df.index, flu_b, bottom=ah1n1 + ah3n2 + bunsubtyped, label='Influenza B', color='cornflowerblue')

        ax2 = ax1.twinx()
        ax2.plot(df.index, perc_a, label='% Positive A', color='darkred', linewidth=2)
        ax2.plot(df.index, perc_b, label='% Positive B', color='darkblue', linewidth=2)

        ax1.set_title("Strain Composition & Positivity Rate", fontsize=16)
        ax1.set_xlabel("Surveillance Week", fontsize=16)
        ax1.set_ylabel("Positive Tests", fontsize=14)
        ax2.set_ylabel("% Tests Positive", fontsize=14)
        ax1.set_xticks(df.index[::2])
        ax1.set_xticklabels(df['Surveilaince Week'].iloc[::2], rotation=45, ha='right')
        ax1.legend(loc='upper left')
        ax2.legend(loc='upper right')
        st.pyplot(fig1)

    # --- Boxplot ---
    with col2:
        st.markdown("### Distribution of Cases by Type")
        fig2, ax_box = plt.subplots(figsize=(12, 6), dpi=150)
        df_box = pd.DataFrame({
            'A(H1N1)': ah1n1,
            'A(H3N2)': ah3n2,
            'A(Unsubtyped)': bunsubtyped,
            'Influenza B': flu_b
        })
        df_box.plot(kind='box', ax=ax_box, patch_artist=True)
        ax_box.set_title("Boxplot of Influenza Cases (2023–2024)", fontsize=16)
        ax_box.set_xlabel("Influenza Strain", fontsize=16)
        ax_box.set_ylabel("Number of Cases", fontsize=14)
        ax_box.tick_params(axis='both', labelsize=14)
        ax_box.grid(False)
        st.pyplot(fig2)

    # --- Fourier + STL + SARIMA Modeling ---
    st.markdown("### Influenza Total Forecasting")
    df['Cases'] = influenza_total

    df_weekly = df['Cases'].asfreq('W-SAT')
    df_weekly = df_weekly.interpolate(method='linear')

    def add_fourier_terms(df, period=52, order=2):
        t = np.arange(len(df))
        for k in range(1, order + 1):
            df[f'sin_{k}'] = np.sin(2 * np.pi * k * t / period)
            df[f'cos_{k}'] = np.cos(2 * np.pi * k * t / period)
        return df

    df_fourier = df_weekly.to_frame(name='Cases')
    df_fourier = add_fourier_terms(df_fourier.copy(), period=52, order=2)
    X = sm.add_constant(df_fourier[[col for col in df_fourier.columns if 'sin' in col or 'cos' in col]])
    y = df_fourier['Cases']
    model = sm.OLS(y, X).fit()

    pred_train = model.get_prediction(X)
    pred_train_summary = pred_train.summary_frame(alpha=0.05)
    df_fourier['Fitted'] = pred_train_summary['mean']
    df_fourier['Lower'] = pred_train_summary['mean_ci_lower']
    df_fourier['Upper'] = pred_train_summary['mean_ci_upper']

    # STL decomposition
    df_log = np.log(df_weekly + 1e-6)
    stl_log = STL(df_log, period=52)
    result_log = stl_log.fit()
    stl_fitted = np.exp(result_log.trend + result_log.seasonal)

    # SARIMA model
    sarima_model = SARIMAX(df_weekly, order=(1, 1, 1), seasonal_order=(1, 1, 1, 52))
    sarima_result = sarima_model.fit(disp=False)
    fitted_sarima = sarima_result.fittedvalues

    fig_total, ax_total = plt.subplots(figsize=(16, 9))
    ax_total.plot(df_weekly.index, df_weekly, label='Actual Cases', color='black', marker='o')
    ax_total.plot(df_fourier.index, df_fourier['Fitted'], label='Fourier Fit', color='red')
    ax_total.plot(df_weekly.index, stl_fitted, label='STL Fit', color='green')
    ax_total.plot(df_weekly.index, fitted_sarima, label='SARIMA Fit', color='blue', linestyle='--')
    ax_total.fill_between(df_fourier.index, df_fourier['Lower'], df_fourier['Upper'], color='red', alpha=0.2)
    ax_total.set_xticks(df.index[::2])
    ax_total.set_xticklabels(df['Surveilaince Week'].iloc[::2], rotation=45, ha='right')
    ax_total.set_title('Fourier, STL, and SARIMA Fitting', fontsize=18)
    ax_total.set_xlabel('Surveillance Week', fontsize=14)
    ax_total.set_ylabel('Cases', fontsize=14)
    ax_total.legend()
    ax_total.grid(True)
    plt.tight_layout()
    st.pyplot(fig_total)

    # --- SEIRV MODEL ---
    st.markdown("### SEIRV Model Forecasting (CmdStanPy)")

    @st.cache_resource
    def load_seirv_model():
        return CmdStanModel(stan_file="seirv_model.stan")

    @st.cache_data
    def run_seirv_model(model, observed_cases):
        N = 40097761
        initial_state = [N - 110, 0, 0, 110, 0, 0]
        params = [0.000000053, 0.33, 0.0004, 0.1, 0.001, 0.2, 0.14]
        T = len(observed_cases)
        time = np.arange(1, T + 1)

        stan_data = {
            "T": T,
            "ts": time,
            "t0": 0 - 1e-6,
            "initial_state": initial_state,
            "parameter": params,
            "incidence": observed_cases
        }
        return model.sample(data=stan_data, chains=4, iter_sampling=1000, iter_warmup=500, seed=123)

    if os.path.exists("seirv_model.stan"):
        seirv_model = load_seirv_model()
        with st.spinner("Running SEIRV model..."):
            seirv_fit = run_seirv_model(seirv_model, df['Influenza Total'].values)

        pred_cases = seirv_fit.stan_variable("pred_cases")
        median_pred = np.median(pred_cases, axis=0)
        ci95_low = np.percentile(pred_cases, 2.5, axis=0)
        ci95_high = np.percentile(pred_cases, 97.5, axis=0)
        ci50_low = np.percentile(pred_cases, 25, axis=0)
        ci50_high = np.percentile(pred_cases, 75, axis=0)

        fig_seirv, ax_seirv = plt.subplots(figsize=(16, 9))
        ax_seirv.fill_between(df.index, ci95_low, ci95_high, color='lightblue', alpha=0.3, label="95% CI")
        ax_seirv.fill_between(df.index, ci50_low, ci50_high, color='deepskyblue', alpha=0.5, label="50% CI")
        ax_seirv.plot(df.index, median_pred, label="Predicted Median", color='blue', linewidth=2)
        ax_seirv.scatter(df.index, df['Influenza Total'], label="Observed Cases", color='black', s=40, zorder=5)

        ax_seirv.set_xticks(df.index[::2])
        ax_seirv.set_xticklabels(df['Surveilaince Week'].iloc[::2], rotation=45, ha='right')
        ax_seirv.set_title("Observed vs Predicted Influenza Cases (SEIRV Model)", fontsize=18)
        ax_seirv.set_xlabel("Surveillance Week", fontsize=14)
        ax_seirv.set_ylabel("Number of Cases", fontsize=14)
        ax_seirv.legend()
        ax_seirv.grid(True)
        plt.tight_layout()
        st.pyplot(fig_seirv)
    else:
        st.error("SEIRV Stan model file ('seirv_model.stan') not found. Please upload it.")

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.statespace.sarimax import SARIMAX
from prophet import Prophet
from statsmodels.tsa.forecasting.stl import STLForecast
from statsmodels.tsa.arima.model import ARIMA
import streamlit as st

st.set_page_config(layout="wide")
st.title("\U0001F9A0 Influenza in Canada")

# Sidebar controls
st.sidebar.header("Filter")
# date_range = st.sidebar.date_input("Select Date Range")
region = st.sidebar.selectbox("Region", ["Canada (National)", "Ontario", "Quebec", "BC", "Alberta"])
n_weeks = st.sidebar.slider("Weeks to Forecast", min_value=4, max_value=52, value=26)

# File upload
uploaded_file = st.sidebar.file_uploader("Upload CSV (with 'Date', 'Virus', 'Cases')", type='csv')

def add_fourier_terms(df, period=52, order=2, start_t=0):
    t = np.arange(start_t, start_t + len(df))
    for k in range(1, order + 1):
        df[f'sin_{k}'] = np.sin(2 * np.pi * k * t / period)
        df[f'cos_{k}'] = np.cos(2 * np.pi * k * t / period)
    df['time'] = t
    return df

if uploaded_file:
    df = pd.read_csv(uploaded_file)
    df.rename(columns={"Number of detections": "Cases", "Week end date": "Date"}, inplace=True)
    df['Date'] = pd.to_datetime(df['Date'])
    df = df[df['Virus'] == 'Influenza']
    df.set_index('Date', inplace=True)
    df['Cases'] = pd.to_numeric(df['Cases'], errors='coerce')
    df = df.dropna(subset=['Cases'])
    df_weekly = df['Cases'].asfreq('W-SAT').interpolate()

    # KPIs
    cumulative_cases = int(df_weekly.sum())
    season_cases = int(df_weekly[-52:].sum())
    weekly_cases = int(df_weekly[-1])
    vaccinations = 3456  # placeholder

    kpi1, kpi2, kpi3, kpi4 = st.columns(4)
    kpi1.metric("Cumulative Cases", f"{cumulative_cases:,}")
    kpi2.metric("New Cases (This Season)", f"{season_cases:,}")
    kpi3.metric("New Cases (This Week)", f"{weekly_cases:,}")
    kpi4.metric("Vaccinalizations", f"{vaccinations:,}")

    # Forecast Models
    df_fourier = df_weekly.to_frame(name='Cases')
    df_fourier = add_fourier_terms(df_fourier, period=52, order=2)
    X_cols = [col for col in df_fourier.columns if 'sin' in col or 'cos' in col] + ['time']
    df_fourier[X_cols] = df_fourier[X_cols].apply(pd.to_numeric, errors='coerce')
    X = sm.add_constant(df_fourier[X_cols])
    y = df_fourier['Cases']
    model = sm.OLS(y, X).fit()
    df_fourier['Fitted'] = model.predict(X)

    future_dates = pd.date_range(df_weekly.index[-1] + pd.Timedelta(weeks=1), periods=n_weeks, freq='W-SAT')
    df_future = pd.DataFrame(index=future_dates)
    df_future = add_fourier_terms(df_future, period=52, order=2, start_t=len(df_fourier))
    X_future = sm.add_constant(df_future[X_cols])
    df_future['Fourier'] = model.predict(X_future)

    sarima_model = SARIMAX(df_weekly, order=(1,1,1), seasonal_order=(1,1,1,52))
    sarima_result = sarima_model.fit(disp=False)
    sarima_forecast = sarima_result.get_forecast(steps=n_weeks).predicted_mean

    df_prophet = df_weekly.reset_index()
    df_prophet.columns = ['ds', 'y']
    prophet_model = Prophet(weekly_seasonality=True)
    prophet_model.fit(df_prophet)
    future = prophet_model.make_future_dataframe(periods=n_weeks, freq='W-SAT')
    forecast = prophet_model.predict(future)
    prophet_forecast = forecast.set_index('ds').loc[future_dates, 'yhat']

    stl_forecast = STLForecast(df_weekly, ARIMA, model_kwargs={"order": (1,1,1)}, period=52)
    stl_result = stl_forecast.fit()
    stl_forecast_values = stl_result.forecast(n_weeks)

    forecast_df = pd.DataFrame(index=future_dates)
    forecast_df['Fourier'] = df_future['Fourier']
    forecast_df['STL'] = stl_forecast_values.values
    forecast_df['SARIMA'] = sarima_forecast.values
    forecast_df['Prophet'] = prophet_forecast.values

    # Layout with charts
    row1, row2 = st.columns(2)

    with row1:
        st.subheader("Weekly Influenza Cases Over Time")
        fig1, ax1 = plt.subplots()
        ax1.plot(df_weekly, label='Observed')
        ax1.set_title("Observed Weekly Cases")
        ax1.grid(True)
        st.pyplot(fig1)

    with row2:
        st.subheader("Forecasting Flu Cases")
        method = st.selectbox("Forecasting Method", forecast_df.columns)
        fig2, ax2 = plt.subplots()
        ax2.plot(df_weekly, label='Observed')
        ax2.plot(forecast_df.index, forecast_df[method], label=f"{method} Forecast", linestyle='--')
        ax2.axvline(df_weekly.index[-1], color='gray', linestyle='--')
        ax2.legend()
        ax2.set_title("Forecasted Flu Cases")
        ax2.grid(True)
        st.pyplot(fig2)

    st.subheader("[Placeholder] Virus Strain Composition")
    st.image("https://via.placeholder.com/600x200?text=Strain+Composition+Chart")

    st.subheader("[Placeholder] Healthcare Utilization")
    st.image("https://via.placeholder.com/600x200?text=ICU+vs+Hosp+vs+ER")

    st.subheader("[Placeholder] Map of Canada")
    st.image("https://via.placeholder.com/400x250?text=Map+of+Canada")
```
